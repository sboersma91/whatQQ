```python
def clear_board(board):
    for key in board.keys():
        board[key]= ' '
    return board
```

```python
def board_view():
    print(board['7'] + '|' + board['8'] + '|' + board['9'])
    print('-+-+-')
    print(board['4'] + '|' + board['5'] + '|' + board['6'])
    print('-+-+-')
    print(board['1'] + '|' + board['2'] + '|' + board['3'])
```

```python
def person_play(num, xo):
    if num == 1:
        board['1']=xo
    elif num == 2:    
        board['2']=xo
    elif num == 3:
        board['3']=xo
    elif num == 4:
        board['4']=xo
    elif num == 5:
        board['5']=xo
    elif num == 6:
        board['6']=xo
    elif num == 7:
        board['7']=xo
    elif num == 8:
        board['8']=xo
    elif num == 9:
        board['9']=xo
    return board
```

```python
def who_wins(board, xo):
    if  board['1'] == board['2'] == board['3'] == xo: #Horiz
        return xo
    elif board['4'] == board['5'] == board['6'] == xo: #Horiz
        return xo
    elif board['7'] == board['8'] == board['9'] == xo: #Horiz
        return xo
    elif board['1'] == board['5'] == board['9'] == xo: #diagonal
        return xo
    elif board['3'] == board['5'] == board['7'] == xo: #diagonal
        return xo
    elif board['1'] == board['4'] == board['7'] == xo: #Vert
        return xo
    elif board['2'] == board['5'] == board['8'] == xo: #Vert
        return xo
    elif board['3'] == board['6'] == board['9'] == xo: #Vert
        return xo
    else:
        pass
```
